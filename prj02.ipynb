# Introduction:
This project runs two separate pieces of code that statistically analyze and classify a dataset containing 42,305 different songs from Spotify. The Goal is to create different visualizations and provide numeric analysis for the attributes of each song. This includes the distribution of each attribute per genre as well as multiple sub categories of analysis. The attributes include:
* danceability
* energy 
* key 
* loudness 
* speechiness 
* acousticness 
* instrumentalness 
* liveness 
* valence 
* tempo 
* duration_ms

The data set also includes 15 different genres. The classification part of this project, is to try and guess the genre of the song based on the values of these different attributes. The dataset can be accessed here:
https://www.kaggle.com/mrmorj/dataset-of-songs-in-spotify

# Dataset:
With this data set having been collected by a third party, there was multiple elements that we did not need for our analysis. The attributes that were kept are strictly numerical and calculated by Spotify for every song in their databse, not just this sample we are working with. We also kept the genre of each song for practical purposes, but this is the only section of the data that is not numerical. Everything else including song names and artist names that were irrelevant to the analysis were dropped. A further eplanation of how Spotify determines the values for these attributes can be found here: https://developer.spotify.com/documentation/web-api/reference/#/operations/get-several-audio-features

# Preprocessing:
In order to properly do the analysis, multiple columns had to be dropped. The very first step in processing the data included importing the csv into a Pandas DataFrame. Once this was done, the unneeded columns were dropped leaving only the attributes listed above, and then the genres column to be used as the target class later, as well as an identifier for the statistical analysis portion. After that, the DataFrame was converted into a NumPy array to facilitate the analysis. The Numpy array was then split into 11 sub-arrays, each seperated by the distinct attributes to analyze. These sub-arrays were then further split up into genre_attribute arrays as the final set of arrays that would be later populated accordingly.

# Classification:
Three distinct models were used to predict the genre of each song based on the attributes. In order to save time and keep the data consistent, only 20% of the dataset was used using the exact same training and testing DataFrames for the three models. Here is the code and results for a Decision Tree Classifier, Naive Bayes Classifier, and Artificial Neural Network Classifier respectively. 

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
import pandas as pd
import os, random
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier 

# definitions/parameters
DATA_FOLDER = '../data'
CSV_FILE_NAME   = 'genres_v2.csv'

if __name__ == "__main__":

    # Generate random to identify rows to be tested
    random.seed(0)
    # Import data from file
    df = pd.read_csv(os.path.join(DATA_FOLDER, CSV_FILE_NAME), low_memory = False)

    # Drop all columns except ones to be tested for and target (genre)
    df = df[['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'genre']]

    # Keep for testing
    # print(df)

    # Select rows to train randmoly, 20% sample
    rows_train = random.sample(range(len(df)), k=int(len(df) * .2))

    # Split training and testing rows into separate DFs
    df_train = pd.DataFrame(columns=df.columns)
    df_test = pd.DataFrame(columns=df.columns)
    for i, row in df.iterrows():
        if i in rows_train:
            df_train = df_train.append(row)
        else:
            df_test = df_test.append(row)
    
    # Decision Tree Classifier
    print('Training Decision Tree Classifier...')
    # Train the set
    X = df_train.iloc[:,0:-1].values
    Y = df_train.iloc[:,-1].values
    model = DecisionTreeClassifier().fit(X, Y)

    correct = 0
    for _, row in df_test.iterrows():
        value_predict = model.predict([row[:-1].values])
        if value_predict == row['genre']:
            correct += 1
    print('accuracy:', correct/len(df_test.index))

    # --------------------------------------------------------------------
    # Naive Bayes Classifier
    print('Training Naive Bayes Classifier...')
    gnb = GaussianNB()
    gnb.fit(X,Y)
    correct = 0
    for _, row in df_test.iterrows():
        value_predict = gnb.predict([row[:-1].values])
        if value_predict == row['genre']:
            correct += 1
    print('accuracy:', correct/len(df_test.index))

    # --------------------------------------------------------------------
    # ANN Classifier
    print('Training ANN Classifier...')
    clf = MLPClassifier(hidden_layer_sizes=(10,5), max_iter=3000, random_state=0)
    clf.fit(X, Y)
    correct = 0
    for _, row in df_test.iterrows():
        value_predict = clf.predict([row[:-1].values])
        if value_predict == row['genre']:
            correct += 1
    print('accuracy:', correct/len(df_test.index))

    
```
